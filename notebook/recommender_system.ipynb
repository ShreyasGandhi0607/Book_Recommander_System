# Popularity Based Recommendation System

```python
ratings
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.head(5)
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')
```

```python
ratings_with_name = ratings_with_name.dropna(subset=['Book-Rating'])
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title', as_index=False)['Book-Rating'].mean()
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df,on=['Book-Title'])
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popularity_df = popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

# Collaborative Based Filtering Approach 

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_log = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_log)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

#### Now below are the books which have more than 50 ratings and users who have rated more than 200 books 

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.shape
# 706 books
# 810 users
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 706 books ka eucledian distance nikala hai 706 books ke sath
similarity_score = cosine_similarity(pt)
```

```python
# similarity_score[0]
```

### Recommender System

```python
def recommand(book_name):
    """recommandation function
    Keyword arguments:
    input -- book name
    Return: suggestions
    """
    # fetch index from book_name
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x: x[1],reverse=True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
    # return suggestions
    
```

```python
recommand('Message in a Bottle')
```

```python
recommand("The Notebook")
```

```python
recommand("The Da Vinci Code")
```
